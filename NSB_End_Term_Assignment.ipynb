```python
# Query 9: Find Documents Containing Specific Keywords in the Abstract Using Full-Text Search
# Create a text index on the 'abstract' field (run this once)
collection.create_index([("abstract", pymongo.TEXT)])

# Now run the query for documents containing specific keywords
results = collection.find({"$text": {"$search": "diphoton cross section"}})
df9 = pd.DataFrame(list(results))
df9
```

```python
# Query 10: Aggregation Pipeline to Find the Average Number of Versions per Document
pipeline = [
    {
        "$project": {
            "num_versions": {
                "$cond": [
                    {"$isArray": "$versions"}, # Check if 'versions' is an array
                    {"$size": "$versions"},     # If true, get the size of the array
                    0                           # If false, return 0
                ]
            }
        }
    },
    {"$group": {"_id": None, "average_versions": {"$avg": "$num_versions"}}}
]

results = collection.aggregate(pipeline)
df10 = pd.DataFrame(results)
df10
```

```python
# Query 11: Aggregation Pipeline to Calculate the Average Number of Authors per Document
pipeline = [
    # Step 1: Project the size of the authors_parsed array if it exists and is an array, otherwise 0
    {"$project": {"num_authors": {"$cond": [
                                         {"$and": [{"$isArray": "$authors_parsed"}, {"$ne": ["$authors_parsed", []]} ]},
                                         {"$size": "$authors_parsed"},
                                         0]}}},

    # Step 2: Group all documents and calculate the average number of authors
    {"$group": {"_id": None, "average_authors": {"$avg": "$num_authors"}}}
]

results = collection.aggregate(pipeline)
df11 = pd.DataFrame(results)
df11
```

```python
# Query 12: Find Documents Where a Field is Missing or Null (Optional Fields)
results = collection.find({"$or": [{"license": {"$exists": False}},
                                  {"license": None}]})

df12 = pd.DataFrame(list(results))
df12
```

```python
# Query 13: Aggregation Pipeline to Find the Most Frequent Words in Titles
pipeline = [
    # Step 1: Tokenize the title field into individual words
    {"$project": {"title_words": {"$split": ["$title", " "]}}},

    # Step 2: Unwind the array of words to have one document per word
    {"$unwind": "$title_words"},

    # Step 3: Group by the word and count occurrences
    {"$group": {"_id": "$title_words", "count": {"$sum": 1}}},

    # Step 4: Sort by count in descending order and limit to top 10 words
    {"$sort": {"count": -1}},
    {"$limit": 10}
]

results = collection.aggregate(pipeline)
df13 = pd.DataFrame(results)
df13
```

```python
# Query 14: Aggregation Pipeline to Calculate the Distribution of Documents by Year
pipeline = [
    # Step 1: Convert update_date to date object if it is a string
    {"$addFields": {
        "update_date": {
            "$cond": [
                {"$eq": [{"$type": "$update_date"}, "string"]},
                {"$toDate": "$update_date"},
                "$update_date"
            ]
        }
    }
    },

    # Step 2: Project the year from the update_date field
    {"$project": {"year": {"$year": "$update_date"}}},

    # Step 3: Group by year and count the number of documents per year
    {"$group": {"_id": "$year", "count": {"$sum": 1}}},

    # Step 4: Sort by year in ascending order
    {"$sort": {"_id": 1}}
]

results = collection.aggregate(pipeline)
df14 = pd.DataFrame(results)
df14
```

```python
# Query 15: Aggregation Pipeline to Find Documents with the Longest Titles
pipeline = [
    # Step 1: Project the length of the title field if it exists, otherwise 0
    {"$project": {"title": 1, "title_length": {"$cond": [
                                                        {"$ifNull": ["$title", False]},
                                                        {"$strLenCP": "$title"},
                                                        0]}}},

    # Step 2: Sort by title length in descending order
    {"$sort": {"title_length": -1}},

    # Step 3: Limit to top 5 documents with the longest titles
    {"$limit": 5}
]

results = collection.aggregate(pipeline)
df15 = pd.DataFrame(results)
df15
```
